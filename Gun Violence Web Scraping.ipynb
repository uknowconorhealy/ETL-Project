```python
import pandas as pd
```

```python
# Get url for table
url = 'https://en.wikipedia.org/wiki/Gun_violence_in_the_United_States_by_state'
```

```python
#Bring table into pandas
tables = pd.read_html(url)
tables
```

```python
type(tables)
```

```python
# Convert table to pandas dataframe and add column titles
gun_df = tables[1]
gun_df.columns = ['State', 'Total Population', 'Murders and Nonnegligent Manslaughter', 'Murders', 
                  'Gun Murders', 'Gun Ownership (%)', 'Murders and Nonnegligent Manslaughter Rate (%)', 'Murder Rate (%)', 'Gun Murder Rate (%)']
gun_df.head()
```

```python
#Drop extra row
gun_df = gun_df.iloc[1:]
gun_df.head()
```

```python
#Set State as index
gun_df.set_index('State', inplace = True)
gun_df.head()
```

```python

#Clean up records with additions
gun_df['Murders'] = gun_df['Murders'].map(lambda x: x.rstrip('[a]').rstrip('[b]'))
gun_df['Gun Murders'] = gun_df['Gun Murders'].map(lambda x: x.rstrip('[a]').rstrip('[b]'))
gun_df['Murder Rate (%)'] = gun_df['Murder Rate (%)'].map(lambda x: x.rstrip('[a]').rstrip('[b]'))
gun_df['Gun Murder Rate (%)'] = gun_df['Gun Murder Rate (%)'].map(lambda x: x.rstrip('[a]').rstrip('[b]'))

#Replace — with NaN 
gun_df['Murders'] = gun_df['Murders'].replace({'—':'NULL'})
gun_df['Gun Murders'] = gun_df['Gun Murders'].replace({'—':'NULL'})
gun_df['Murder Rate (%)'] = gun_df['Murder Rate (%)'].replace({'—':'NULL'})
gun_df['Gun Murder Rate (%)'] = gun_df['Gun Murder Rate (%)'].replace({'—':'NULL'})
gun_df.head(20)
```

```python
# Write to CSV for uploading to SQL

gun_df.to_csv('resources/gun_murder_rates.csv',index=True,header = True)
```
