```python
import pandas as pd
```

```python
csv = 'resources/weapon_type.csv'
```

```python
weapon_df = pd.read_csv(csv, encoding="ISO-8859-1")
weapon_df.head()
```

```python
#Add columns and delete uneccessary rows
weapon_df.columns = ['State', 'Total Murders', 'Total Firearms', 'Handguns', 'Rifles', 'Shotguns', 'Unknown Firearms',
                    'Knives', 'Other Weapons', 'Hands/Feet/etc']
weapon_df = weapon_df.iloc[3:54]

weapon_df.head(100)
```

```python
#Clean up extra parts of State
weapon_df['State'] = weapon_df['State'].map(lambda x: x.rstrip('3'))
weapon_df.head()
                                            
                                            
```

```python
#Set State as index
weapon_df.set_index('State', inplace = True)
weapon_df.head()
```

```python
#Bring in Gun Violence CSV
gun_violence_df = pd.read_csv('resources/gun_murder_rates.csv', encoding="ISO-8859-1")
gun_violence_df.head()
```

```python
#Merge columns
merge_gun_df = pd.merge(weapon_df, gun_violence_df, on="State", how="left")
merge_gun_df.head()
```

```python
#Delete redundant columns and organize
del merge_gun_df['Murders']
del merge_gun_df['Gun Murders']
merge_gun_df.set_index('State', inplace = True)
merge_gun_df.head()
```

```python
# Write to CSV for uploading to SQL
merge_gun_df.to_csv('resources/state_violence.csv',index=True,header = True)
```

```python
#Set Up SQL Alchemy For FlaskApp
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
#Create SQLite Database
engine = create_engine("sqlite:///resources/state_violence.sqlite")

merge_gun_df.to_sql('type_and_rate', engine, if_exists='replace', index = True)
```
